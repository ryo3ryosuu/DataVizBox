## Functions  

```python
def contents(df):
    aa = pd.DataFrame(df.dtypes)
    cont = pd.DataFrame({'Name': aa.index,
                         'Type': aa.iloc[:,0]}).reset_index(drop=True)
    return(pd.DataFrame(cont))
```
